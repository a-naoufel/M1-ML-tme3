```python
from tme3 import *
uspsdatatrain = "../data/USPS_train.txt"
uspsdatatest = "../data/USPS_test.txt"
alltrainx,alltrainy = load_usps(uspsdatatrain)
alltestx,alltesty = load_usps(uspsdatatest)
neg = 5
pos = 6
datax,datay = get_usps([neg,pos],alltrainx,alltrainy)
testx,testy = get_usps([neg,pos],alltestx,alltesty)
```

```python
import numpy as np

y_train = np.where(alltrainy == 6, 1, -1)
y_test  = np.where(alltesty  == 6, 1, -1)

X_train = alltrainx.reshape(alltrainx.shape[0], -1) / 255.0
X_test  = alltestx.reshape(alltestx.shape[0], -1) / 255.0
```

```python
X_train_b = np.c_[np.ones(X_train.shape[0]), X_train]
X_test_b  = np.c_[np.ones(X_test.shape[0]),  X_test]

m = Lineaire(perceptron_loss, perceptron_grad, max_iter=50, eps=0.1)
m.fit(X_train_b, y_train, testx=X_test_b, testy=y_test, batch_size=32)

print("perceptron - train acc:", m.score(X_train_b, y_train))
print("perceptron - test  acc:", m.score(X_test_b,  y_test))
```

```python
from sklearn.linear_model import Perceptron

skp = Perceptron(
    max_iter=50,          # epochs
    eta0=0.1,             # pas
    shuffle=True,
    random_state=0,
    tol=None              # important: sinon arrêt anticipé
)
skp.fit(X_train, y_train)

print("sklearn perceptron - train acc:", skp.score(X_train, y_train))
print("sklearn perceptron - test  acc:", skp.score(X_test,  y_test))
```
